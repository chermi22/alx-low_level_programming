```python
import pdfplumber
import re

documents = []
filenames = []

for filename in uploaded.keys():
    text = ""
    with pdfplumber.open(filename) as pdf:
        for page in pdf.pages:
            page_text = page.extract_text()
            if page_text:
                text += page_text + " "

    # Clean text
    text = re.sub(r'\n+', ' ', text)
    text = text.lower()

    documents.append(text)
    filenames.append(filename)
```

```python
chunks = []
chunk_size = 500  # approx words per chunk

for i, doc in enumerate(documents):
    words = doc.split()
    for j in range(0, len(words), chunk_size):
        chunk_text = " ".join(words[j:j+chunk_size])
        chunks.append({"filename": filenames[i], "text": chunk_text})
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

model = SentenceTransformer('all-MiniLM-L6-v2')

# Create embeddings
texts = [chunk['text'] for chunk in chunks]
embeddings = model.encode(texts, show_progress_bar=True)
embeddings = np.array(embeddings)  # convert to numpy array
```

```python
import faiss

dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(embeddings)

print(f"FAISS index created with {index.ntotal} vectors")
```

```python
def search(query, top_k=3):
    query_vec = model.encode([query])
    distances, indices = index.search(np.array(query_vec), top_k)

    results = []
    for idx in indices[0]:
        results.append(chunks[idx]["text"])
    return results

# Example query
query = "Explain the course objectives"
results = search(query)
for i, res in enumerate(results, 1):
    print(f"Result {i}:\n{res}\n")
```

```python
from transformers import pipeline

# Use a small summarization model for Colab
summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")
```

```python
def chatbot_summarized(query, top_k=3, max_length=150):
    relevant_chunks = search(query, top_k=top_k)

    # Combine chunks into a single text
    combined_text = " ".join(relevant_chunks)

    # Summarize
    summary = summarizer(combined_text, max_length=max_length, min_length=30, do_sample=False)[0]['summary_text']

    print("\n--- Chatbot Answer ---\n")
    print(summary)
    print("----------------------\n")

# Interactive loop
while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot session ended.")
        break
    chatbot_summarized(user_input, top_k=3, max_length=150)
```

```python
def safe_chatbot(query, top_k=3, max_length=150):
    # Check if any chunks exist
    if len(chunks) == 0:
        print("❌ No chunks available. Upload PDFs and generate embeddings first.")
        return

    # Limit top_k to number of chunks
    k = min(top_k, len(chunks))

    # Perform search
    relevant_chunks = search(query, top_k=k)

    if len(relevant_chunks) == 0:
        print("❌ No relevant chunks found for your query.")
        return

    # Combine top chunks
    combined_text = " ".join(relevant_chunks)

    # Summarize if possible
    try:
        summary = summarizer(combined_text, max_length=max_length, min_length=30, do_sample=False)[0]['summary_text']
    except:
        summary = combined_text  # fallback if summarization fails

    # Print the chatbot response
    print("\n--- Chatbot Answer ---\n")
    print(summary)
    print("----------------------\n")
```

```python
while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot session ended.")
        break
    safe_chatbot(user_input, top_k=3, max_length=150)
```
